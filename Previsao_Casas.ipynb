##Dataset Casas

* https://www.kaggle.com/competitionshouse-prices-advanced-regression-techniques/data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
```

```python
dados = pd.read_csv('/content/casas.csv', sep=',')
```

```python
dados.head()
```

```python
# colunas = ['tamanho', 'preco']
# dados = dados[colunas]
```

```python
dados.head()
```

```python
x = dados.drop('preco', axis=1)
y = dados['preco']
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, train_size=0.7, random_state=42)
```

```python
modelo  = LinearRegression()
modelo.fit(x_treino, y_treino)
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
# Dados o tamnho a Regressão prevê o valor da casa
modelo.predict([[120,2001,2]])
```

```python
# Serializar o modelo import pickle
pickle.dump(modelo, open('modelo.sav', 'wb'))
```
